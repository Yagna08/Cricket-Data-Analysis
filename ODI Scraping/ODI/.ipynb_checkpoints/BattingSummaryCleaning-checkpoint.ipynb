```python
import pandas as pd
import numpy as np
df = pd.read_csv('BattingSummaryINDIA.csv')
print(df.tail())
```

```python
df = df.drop(['Abc'],axis=1)
print(df)
```

```python
df['Name']=df['Name'].apply(lambda x: x.replace('�',''))
df['Name']=df['Name'].apply(lambda x: x.replace('(c)',''))
df['Name']=df['Name'].apply(lambda x: x.replace('†',''))
df['Name']=df['Name'].str.rstrip()


df.drop(df[df['SR']=='-'].index,inplace=True)
```

```python
df1 = pd.get_dummies(df['Out/NotOut'],drop_first = True)
# print(df1)
```

```python
df = pd.concat([df,df1],axis=1)
# print(df)
```

```python
df.drop(['Out/NotOut'],inplace=True,axis=1)
# print(df)
```

```python
df['Matches Played'] = df.groupby('Name')['Name'].transform('count')
```

```python
df = df.sort_values(by='Name')
# print(df)
```

```python
grouped = df.groupby('Name')
outs = grouped[['Out']].sum().reset_index().drop_duplicates()
# print(outs)
```

```python
df = df.drop_duplicates(subset=['Name'])
# df
```

```python
aggregated = grouped[['Runs','Balls',"4s","6s",]].sum()
SR_aggregated = grouped[['SR']].sum()
aggregated
```

```python
df=df.drop('Out',axis=1)
```

```python
# print(len(df))
# print(type(outs['Out']))
# print(outs['Out'])
# # print(new_outs['Out'])
```

```python
df['Out']=list(outs['Out'])
# df
```

```python
# print(type(aggregated['Runs']))
# print(type(df['Runs']))
df['Runs']=list(aggregated['Runs'])
df['4s']=list(aggregated['4s'])
df['6s']=list(aggregated['6s'])
df['Balls']=list(aggregated['Balls'])
df['SR'] =df['Runs']/df['Balls']*100
df['Avg'] = df['Runs']/df['Out']
# df = df.set_index('Name')
df
```

```python
import numpy as np
df.loc[df['Avg'] == np.inf, 'Avg'] = df[df['Avg'] == np.inf]['Runs']
df['Avg'].fillna(0, inplace=True)
df
```
