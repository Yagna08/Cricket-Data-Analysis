```python
import pandas as pd
import re
df = pd.read_csv('Old/ODImatchSummaryINDIA.csv')
df = df.drop(['Abc'], axis = 1)
df
```

```python
df['Id']=df['Id'].apply(lambda x: x.replace('\n                                    ',' '))
df['Winner']=df['Winner'].apply(lambda x: x.replace('\n                                    ',''))
df['Margin']=df['Margin'].apply(lambda x: x.replace('\r\n                                    ',''))
df['Date']=df['Date'].apply(lambda x: x.replace('\n                                  ',' '))
df['Margin'] = df['Margin'].apply(lambda x: re.sub(r'\s*\([^)]*\)', '', x))
df["Winner"] = df["Winner"].apply(lambda x: x.strip())
df["Margin"] = df["Margin"].apply(lambda x: x.strip())
df
```

```python
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Sri Lanka', 'SL'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Australia','AUS'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('India','IND'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Pakistan','PAK'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Afghanistan','AFG'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Hong Kong','HKG'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Ireland','IRE'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('New Zealand','NZ'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('South Africa','SA'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('England','ENG'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Bangladesh','BAN'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Scotland','SCO'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Namibia','NAM'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('West Indies','WI'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Netherlands','NET'))
df[['Team1', 'Team2', 'Winner']] = df[['Team1', 'Team2', 'Winner']].apply(lambda x: x.str.replace('Zimbabwe','ZIM'))
df
```

```python
df.to_csv('New/NewODImatchSummaryINDIA.csv')
```
