```python
import pandas as pd
import numpy as np
df = pd.read_csv('BattingSummaryINDIA.csv')
print(df.tail())
```

```python
df = df.drop(['Abc'],axis=1)
print(df)
```

```python
df['Name']=df['Name'].apply(lambda x: x.replace('�',''))
df['Name']=df['Name'].apply(lambda x: x.replace('(c)',''))
df['Name']=df['Name'].apply(lambda x: x.replace('†',''))
df['Name']=df['Name'].str.rstrip()
df['Out/NotOut'] = df['Out/NotOut'].map({'Out': 1, 'Not Out': 0})    

df
```

```python
df.drop(df[df['SR']=='-'].index,inplace=True)
df
```

```python
df.to_csv('NewBattingSummary1.csv')
```
