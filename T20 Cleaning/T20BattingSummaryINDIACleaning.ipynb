```python
import pandas as pd
df = pd.read_csv("Old\T20BattingSummaryINDIA.csv", encoding='iso-8859-1', engine='python')
df
```

```python
df = df.drop(['Abc'],axis=1)
df
```

```python
df['Name']=df['Name'].apply(lambda x: x.replace('�',''))
df['Name']=df['Name'].apply(lambda x: x.replace('(c)',''))
df['Name']=df['Name'].apply(lambda x: x.replace('†',''))
df['Name']=df['Name'].apply(lambda x: x.replace('ï¿½',''))

df['Name']=df['Name'].str.rstrip()
df['Name']=df['Name'].apply(lambda x: x.replace('\r\n                                        ',' '))
df.drop(df[df['SR']=='-'].index,inplace=True)
df
```

```python
df['Out/NotOut']=df['Out/NotOut'].apply(lambda x: x.replace('Not Out','0'))
df['Out/NotOut']=df['Out/NotOut'].apply(lambda x: x.replace('Out','1'))

df
```

```python
df['Match_id']=df['Match_id'].apply(lambda x: x.replace('\r\n                                    ',' '))
df
```

```python
df['Match']=df['Match'].apply(lambda x: x.replace('sri lanka','SL'))
df['Match']=df['Match'].apply(lambda x: x.replace('australia','AUS'))
df['Match']=df['Match'].apply(lambda x: x.replace('hong kong','HKG'))
df['Match']=df['Match'].apply(lambda x: x.replace('india','IND'))
df['Match']=df['Match'].apply(lambda x: x.replace('ireland','IRE'))
df['Match']=df['Match'].apply(lambda x: x.replace('pakistan','PAK'))
df['Match']=df['Match'].apply(lambda x: x.replace('afghanistan','AFG'))
df['Match']=df['Match'].apply(lambda x: x.replace('new zealand','NZ'))
df['Match']=df['Match'].apply(lambda x: x.replace('south africa','SA'))
df['Match']=df['Match'].apply(lambda x: x.replace('england','ENG'))
df['Match']=df['Match'].apply(lambda x: x.replace('bangladesh','BAN'))
df['Match']=df['Match'].apply(lambda x: x.replace('scotland','SCO'))
df['Match']=df['Match'].apply(lambda x: x.replace('namibia','NAM'))
df['Match']=df['Match'].apply(lambda x: x.replace('west indies','WI'))
df['Match']=df['Match'].apply(lambda x: x.replace('netherlands','NET'))
df['Match']=df['Match'].apply(lambda x: x.replace('zimbabwe','ZIM'))
df['Match']=df['Match'].str.split().str[:3].str.join(' ')
df
```

```python
df.to_csv('New/NewT20BattingSummaryINDIA.csv')
```
