```python
import pandas as pd
df = pd.read_csv("T20MatchSummaryINDIA.csv")
df
```

```python
df = df.drop(['Abc'],axis=1)
df
```

```python
for i in ('Team1','Team2','Winner'):
    df[i]=df[i].apply(lambda x: x.replace('Sri Lanka','SL'))
    df[i]=df[i].apply(lambda x: x.replace('Australia','AUS'))
    df[i]=df[i].apply(lambda x: x.replace('Hong Kong','HKG'))
    df[i]=df[i].apply(lambda x: x.replace('India','IND'))
    df[i]=df[i].apply(lambda x: x.replace('Ireland','IRE'))
    df[i]=df[i].apply(lambda x: x.replace('Pakistan','PAK'))
    df[i]=df[i].apply(lambda x: x.replace('Afghanistan','AFG'))
    df[i]=df[i].apply(lambda x: x.replace('New Zealand','NZ'))
    df[i]=df[i].apply(lambda x: x.replace('South Africa','SA'))
    df[i]=df[i].apply(lambda x: x.replace('England','ENG'))
    df[i]=df[i].apply(lambda x: x.replace('Bangladesh','BAN'))
    df[i]=df[i].apply(lambda x: x.replace('Scotland','SCO'))
    df[i]=df[i].apply(lambda x: x.replace('Namibia','NAM'))
    df[i]=df[i].apply(lambda x: x.replace('West Indies','WI'))
    df[i]=df[i].apply(lambda x: x.replace('Netherlands','NET'))
    df[i]=df[i].apply(lambda x: x.replace('Zimbabwe','ZIM'))

df
```

```python
df = df.drop(df[(df['Team1'] == '-') & (df['Team2'] == '-')].index)
df
```

```python
df['Margin'] = df['Margin'].str.split('(', expand=True)[0]
df['Margin'] = df['Margin'].str.replace('\n', '')
df['Date'] = df['Date'].str.replace('\n', '')
df['Id'] = df['Id'].str.replace('\n', '')
df
```

```python
df['Id'] = df['Id'].str.replace('\r', '')
df['Id'] = df['Id'].str.replace('                                    ', ' ')
df['Margin'] = df['Margin'].str.replace('                                    ', ' ')
df['Date'] = df['Date'].str.replace('                                  ', ' ')
df['Winner'] = df['Winner'].str.replace('\n                                    ', ' ')

print(df.head(15))
```

```python
df.to_csv('NewT20MatchSummaryINDIA.csv')
```
